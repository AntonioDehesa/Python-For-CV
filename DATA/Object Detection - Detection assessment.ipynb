```python
import numpy as np, cv2, matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('DATA/car_plate.jpg')
```

```python
car_plate_cascade = cv2.CascadeClassifier('DATA/haarcascades/haarcascade_licence_plate_rus_16stages.xml')
```

```python
def display(img):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

```python
display(img)
```

```python
def detect_car_plate(img):
    car_img = img.copy()
    car_rects = car_plate_cascade.detectMultiScale(car_img)
    for (x,y,w,h) in car_rects:
        cv2.rectangle(car_img, (x,y), (x+w, y+h), (0,0,255),4)
    return car_img
```

```python
result = detect_car_plate(img)
```

```python
display(result)
```

```python
def blur_car_plate(img):
    car_img = img.copy()
    blur = img.copy()
    car_rects = car_plate_cascade.detectMultiScale(car_img)
    for (x,y,w,h) in car_rects:
        blur = blur[y:y+h, x:x+w]
        blurred = cv2.medianBlur(blur, 5)
        car_img[y:y+h, x:x+w] = blurred
    return car_img
```

```python
result = blur_car_plate(img)
```

```python
display(result)
```
